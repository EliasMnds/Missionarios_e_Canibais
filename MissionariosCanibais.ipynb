```python
def aprofundamento_iterativo(raiz):

    limite_profundidade = 0
    fim_busca = False #  true quando chegar ao fim da árvore ou alcançar o nó objetivo
    profundidade_atual = 0
    caminho=[]

    while fim_busca != True:

        visitados=[]

        print("Limite = ", limite_profundidade)

        resultado_busca, fim_busca, caminho = busca(raiz, profundidade_atual, limite_profundidade, caminho, visitados)

        if resultado_busca != None : # se a busca retornou um nó objetivo
            return resultado_busca, caminho[::-1]

        # se um resultado ainda não foi encontrado, aumenta o limite de profundidade
        if fim_busca != True:
            print("Aumentando limite\n --------- \n")
            limite_profundidade += 1

        else:
            return None, ['']

def busca (no, profundidade_atual, limite_profundidade, caminho, visitados) :

    if no in visitados : # Ignorar nós já visitados
        return None, True, caminho

    print("Visitando nó ",no.index,"- Margem Direita:","C" + str(no["CD"]) + "M" + str(no["MD"]) + "B" + str(no["B"]))
    visitados.append(no)

    if no["CE"] == 0 and no["ME"]==0 and no["B"]==1: # nó objetivo encontrado
        print("\nObjetivo alcançado!")
        s = "Nó "+ str(no.index)+" - "+"C"+ str(no["CD"]) + "M"+ str(no["MD"]) + "B"+ str(no["B"])
        caminho.append(s)
        return no, True, caminho # nó objetivo encontrado, busca chega ao fim

    if profundidade_atual == limite_profundidade:
        # profundidade máxima atingida
        if  len (no.successors()) > 0 :
            # Aumentando limite da árvore
            return None, False, caminho # objetivo não encontrado , árvore não chegou ao fim ( a busca continua)
        else:
            return None, True, caminho # objetivo não encontrado, árvore chegou ao fim (fim da busca)

    # Visita os nós filhos
    fim_busca = True
    for i in no.successors():
        resultado_busca, fim_busca_recursiva, caminho = busca (i, profundidade_atual + 1, limite_profundidade,
                                                               caminho, visitados)

        if resultado_busca != None: # se o retorno da busca foi um nó objetivo
            s = "Nó "+ str(no.index)+" - "+"C"+ str(no["CD"]) + "M"+ str(no["MD"]) + "B"+ str(no["B"])
            caminho.append(s)
            return resultado_busca, True, caminho

        # se não encontrar o objetivo, a busca só encerra se não houver mais nós para visitar ou expandir
        fim_busca = fim_busca and fim_busca_recursiva

    return None, fim_busca, caminho
```

```python
raiz = g.vs[0]
caminho = []
resultado_busca, caminho = aprofundamento_iterativo(raiz)

if resultado_busca != None:

    print("\nResultado da busca: nó", resultado_busca.index,"- Margem direita: ",
          "C" + str(resultado_busca["CD"])+ "M" + str(resultado_busca["MD"]) + "B" + str(resultado_busca["B"]),"\n",
          "\nCaminho(margem direita): ", caminho )
else:

    print("\nNão encontrado")
```
